```python
# dropping the columns not neccessary
df = df.drop(['Date','City','AQI_Bucket'],axis=1)
df
```

```python
from sklearn.model_selection import train_test_split

# splitting the dataset to train and test sets
X = df.drop(['AQI'],axis=1)
Y = df['AQI'].astype('int')
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42,shuffle=True)
```

```python
# importing libraries for models
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,classification_report,r2_score,precision_score
```

```python
# training and testing model using linear regression
Lin = LinearRegression()
Lin_model = Lin.fit(X_train,Y_train)
Lin_pred = Lin_model.predict(X_test)
Lin_acc = r2_score(Y_test,Lin_pred)
print("ACCURACY WITH LINEAR REGRESSOR: ",Lin_acc)
```

```python
# regression plot for PM2.5 versus AQI
plt.plot(X_test['PM2.5'],Lin_pred,'o',color='green')
m,b = np.polyfit(X_test['PM2.5'],Lin_pred,1)
plt.title("PM2.5 versus AQI regression plot")
plt.xlabel("PM2.5")
plt.ylabel("AQI")
plt.plot(X_test['PM2.5'],m*X_test['PM2.5']+b,color='black')
plt.show()
```

```python
# regression plot for PM10 versus AQI
plt.plot(X_test['PM10'],Lin_pred,'o',color='teal')
m,b = np.polyfit(X_test['PM10'],Lin_pred,1)
plt.title("PM10 versus AQI regression plot")
plt.xlabel("PM10")
plt.ylabel("AQI")
plt.plot(X_test['PM10'],m*X_test['PM10']+b,color='black')
plt.show()
```

```python
# regression plot for NO2 versus AQI
plt.plot(X_test['NO2'],Lin_pred,'o',color='orange')
m,b = np.polyfit(X_test['NO2'],Lin_pred,1)
plt.title("NO2 versus AQI regression plot")
plt.xlabel("NO2")
plt.ylabel("AQI")
plt.plot(X_test['NO2'],m*X_test['NO2']+b,color='black')
plt.show()
```

```python
# regression plot for SO2 versus AQI
plt.plot(X_test['SO2'],Lin_pred,'o',color='plum')
m,b = np.polyfit(X_test['SO2'],Lin_pred,1)
plt.title("SO2 versus AQI regression plot")
plt.xlabel("SO2")
plt.ylabel("AQI")
plt.plot(X_test['SO2'],m*X_test['SO2']+b,color='black')
plt.show()
```

```python
# regression plot for O3 versus AQI
plt.plot(X_test['O3'],Lin_pred,'o',color='peru')
m,b = np.polyfit(X_test['O3'],Lin_pred,1)
plt.title("O3 versus AQI regression plot")
plt.xlabel("O3")
plt.ylabel("AQI")
plt.plot(X_test['O3'],m*X_test['O3']+b,color='black')
plt.show()
```

```python
# regression plot for CO versus AQI
plt.plot(X_test['CO'],Lin_pred,'o',color='olive')
m,b = np.polyfit(X_test['CO'],Lin_pred,1)
plt.title("CO versus AQI regression plot")
plt.xlabel("CO")
plt.ylabel("AQI")
plt.plot(X_test['CO'],m*X_test['CO']+b,color='black')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

# training and testing model using K-Nearest neighbors
knn = KNeighborsRegressor(n_jobs=-1)
knn_model = knn.fit(X_train,Y_train)
knn_pred = knn_model.predict(X_test)
knn_acc = r2_score(Y_test,knn_pred)
print("ACCURACY WITH K-NEAREST NEIGHBORS REGRESSOR: ",knn_acc)
```

```python
from xgboost import XGBRegressor

# training and testing model using XGBoost
xg = XGBRegressor()
xgmodel = xg.fit(X_train,Y_train)
xgpred = xgmodel.predict(X_test)
xgacc = r2_score(Y_test,xgpred)
print("ACCURACY WITH XGBOOST REGRESSOR: ",xgacc)
```
